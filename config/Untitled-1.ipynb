```python
# Problema 2 seccion I
def conad(s,m,n_numeros):
    X=s.copy()
    k=len(s)
    gen=[]
    for _ in range(n_numeros):
        nuevo=(X[-1]=X[-k]) % m
        X.append(nuevo)
        gen.append(nuevo)
    R=[x/m for x in gen]
    return R

s=[25,57,98,1,15,35]
m=100
R=conad(s,m,10)
print(R)
```

```python
# Problema 2 seccion II
import random
import math

def erlang(k,lam):
    prod=1
    for _ in range(k):
        r=random.random()
        prod*=r

    y=-(1/lam)*math.log(prod)
    return y

k=3
lam=3/5

resultados=[erlang(k,lam) for _ in range(5)]

print("Variables de Erlang:")
for i,val in enumerate(resultados,1):
    print(f"Y{i}={val:.4f}")
```

```python
# Problema 2 seccion III
import math

data=resultados
data.sort()

k=3
lam=0.6
n=len(data)

def erlang_cdf(x,k,lam):
    suma=0
    for i in range(k):
        suma+=(math.exp(-lam*x)*(lam*x)**i)/math.factorial(i)
    return 1-suma

D=0
for i in range(n):
    Fo=(i+1) / n
    Fe=erlang_cdf(data[i], k, lam)
    diff=abs(Fo-Fe)
    if diff>D:
        D=diff
print("D =", D)
```
